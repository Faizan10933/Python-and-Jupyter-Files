```python
from tkinter import *
from tkinter import ttk
class App:
    def __init__(self):
        self.root=Tk()
        self.root.geometry('500x600')
        self.root.resizable(False,False)
        self.root.title('app')
        self.root['bg']='white'
        Label(self.root,text='EMI CALCULATOR',font=('verdana',40),bg='black',fg='yellow').grid(row=0,column=0,columnspan=3,padx=10,pady=10)
        L=['Select Company','Select Model','Price']
        company=['--select--','Samsung','Realme','Redmi','Micromax','Oppo','Vivo','Apple']
        self.d={'Samsung':['J2','J3','J5','J6','J7','A31','A51'],'Realme':['X2','X2 Pro','C2 pro','C3'],'Redmi':['Redmi 8','Redmi Note 8','Redmi 7 pro'],'Mircomax':['Bolt'],'Oppo':['Reno 2','Reno 4'],'Vivo':['V7','V5'],'Apple':['XS','X']}
        self.price={'J2':7000,'J3':12000,'J5':15000,'J6':16000,'J7':17560,'A31':13240,'A51':18000,'X2':23000,'X2 Pro':26000}
        j=1
        for i in L:
            Label(self.root,text=i,font=('arial',12),bg='black',fg='white',width=15).grid(row=j,column=0,padx=10,pady=10,sticky='W')
            Label(self.root,text=':',font=('arial',12),bg='black',fg='white',width=15).grid(row=j,column=1,padx=10,pady=10,sticky='W')
            j+=1
        self.c1=ttk.Combobox(self.root,values=company)
        self.c1.current(0)
        self.c1.grid(row=1,column=2)
        self.c1.bind("<<ComboboxSelected>>",self.getModel)
        self.b1=Button(self.root,text='EMI',fg='white',bg='red',command=self.newwindow)
        self.b1.grid(row=4,column=2)
        self.root.mainloop()
    
    def getModel(self,evt):
        self.company=self.c1.get()
        self.model=ttk.Combobox(self.root,value=self.d[self.company])
        self.model.current(0)
        self.model.bind("<<ComboboxSelected>>",self.getPrice)
        self.model.grid(row=2,column=2)
    def getPrice(self,evt):
        self.m=self.model.get()
        self.rate=self.price[self.m]
        
        self.t1=Entry(self.root)
        self.t1.insert(0,self.rate)
        self.t1.grid(row=3,column=2)
        
        
        
    def newwindow(self):
       # f=open('price.txt','w')
       # f.write(str(self.rate))
       # f.close()
        f1=open('price.txt','r')
        p=f1.read()
        f1.close()
        self.root1=Toplevel(self.root)
        self.root1.geometry('500x500')
        Label(self.root1,text='Get EMI Assitance',font=('arial',20)).grid(row=0,column=0)
        self.dp=Scale(self.root1,from_=1,to=p,orient='horizontal',length=500,font=('arial',12),troughcolor='yellow').grid(row=1,column=0)
        
        
        
    
obj=App()
    
        
```
