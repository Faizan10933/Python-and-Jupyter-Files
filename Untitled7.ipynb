```python
from tkinter import *
import tkinter.scrolledtext as st
from tkinter import filedialog
from tkinter import messagebox
from tkinter import colorchooser
root=Tk()
height=300
width=800

screen_width=root.winfo_screenwidth()
screen_height=root.winfo_screenheight()
x=(screen_width//2)-(width//2)
y=(screen_height//2)-(height//2)
root.geometry("{}x{}+{}+{}".format(width,height,x,y))
root.iconbitmap('notepad.ico')
root.title('untitled notepad')
root.bind('<F11>',lambda event:root.state('zoomed'))
root.bind('<Escape>',lambda event:root.state('normal'))

def newFile():
    txt.delete(1.0,END)
    root.title('Untitled Notepad')
            
def openFile():
    file=filedialog.askopenfilename()
    ftype=file[-3:]
    if ftype!='txt':
        messagebox.showerror('Notepad','Invald File Type')
        return
    txt.delete(1.0,END)
    f=open(file,'r')
    txt.insert(1.0,f.read())
    f.close()
def saveFile():
    file=filedialog.asksaveasfilename(defaultextension='.txt')
    f=open(file,'w')
    f.write(txt.get(1.0,END))
    f.close()
def selectColor():
    (r,h)=colorchooser.askcolor()
    print(r,h)
    txt.config(fg=h)
    f.close()            
menubar=Menu(root)
file=Menu(menubar,tearoff=0)
file.add_command(label='New',accelerator='Ctrl+N',underline=0,command=newFile)
file.add_command(label='Open',accelerator='Ctrl+O',command=newFile)
file.add_command(label='Save',accelerator='Ctrl+S',command=newFile)
file.add_command(label='Save As')
file.add_separator()
file.add_checkbutton(label='Page Setup',onvalue=1)
file.add_command(label='Print')
file.add_separator()
file.add_command(label='Exit')
                    
edit=Menu(menubar,tearoff=0)
edit.add_command(label='Cut',accelerator='Ctrl+X',underline='0')
edit.add_command(label='Copy',accelerator='Ctrl+C')
edit.add_command(label='Paste',accelerator='Ctrl+V')
                
formatmenu=Menu(menubar,tearoff=0)
formatmenu.add_command(label='WordWrap',underline='0')
formatmenu.add_command(label='Font',underline='3')
formatmenu.add_command(label='Color',underline='0',command=selectColor)
                    
menubar.add_cascade(label='File',menu=file,underline=0)
menubar.add_cascade(label='Edit',menu=edit,underline=0)
menubar.add_cascade(label='Format',menu=formatmenu,underline=1)
root.config(menu=menubar)
txt=st.ScrolledText(root,font=('arial',12))
txt.pack(expand=True,fill='both')
root.mainloop()                    
```
